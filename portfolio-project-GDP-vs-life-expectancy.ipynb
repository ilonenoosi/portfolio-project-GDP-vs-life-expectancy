```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#loading data

data = pd.read_csv("all_data.csv")
data.describe()
data.columns.tolist()
```

```python
print(data.head())
```

```python
list_of_all_coutries_life_exp_2015 = []
list_of_all_coutries_life_exp_2000 = []
list_of_countries = ["Chile", "China", "Germany", "USA", "Mexico", "Zimbabwe"]

#chile
chile_data = data[data["Country"] == "Chile"]
chile_2015 = chile_data[chile_data["Year"] == 2015]
chile_2015_life_exp = chile_2015["Life expectancy at birth (years)"].values[0]
chile_2000 = chile_data[chile_data["Year"] == 2000]
chile_2000_life_exp = chile_2000["Life expectancy at birth (years)"].values[0]
list_of_all_coutries_life_exp_2015.append(chile_2015_life_exp)
list_of_all_coutries_life_exp_2000.append(chile_2000_life_exp)

#China
china_data = data[data["Country"] == "China"]
china_2015 = china_data[china_data["Year"] == 2015]
china_2015_life_exp = china_2015["Life expectancy at birth (years)"].values[0]
china_2000 = china_data[china_data["Year"] == 2000]
china_2000_life_exp = china_2000["Life expectancy at birth (years)"].values[0]
list_of_all_coutries_life_exp_2015.append(china_2015_life_exp)
list_of_all_coutries_life_exp_2000.append(china_2000_life_exp)

#Germany
germany_data = data[data["Country"] == "Germany"]
germany_2015 = germany_data[germany_data["Year"] == 2015]
germany_2015_life_exp = germany_2015["Life expectancy at birth (years)"].values[0]
germany_2000 = germany_data[germany_data["Year"] == 2000]
germany_2000_life_exp = germany_2000["Life expectancy at birth (years)"].values[0]
list_of_all_coutries_life_exp_2015.append(germany_2015_life_exp)
list_of_all_coutries_life_exp_2000.append(germany_2000_life_exp)


#USA
usa_data = data[data["Country"] == "United States of America"]
usa_2015 = usa_data[usa_data["Year"] == 2015]
usa_2015_life_exp = usa_2015["Life expectancy at birth (years)"].values[0]
usa_2000 = usa_data[usa_data["Year"] == 2000]
usa_2000_life_exp = usa_2000["Life expectancy at birth (years)"].values[0]
list_of_all_coutries_life_exp_2015.append(usa_2015_life_exp)
list_of_all_coutries_life_exp_2000.append(usa_2000_life_exp)

#Mexico
mexico_data = data[data["Country"] == "Mexico"]
mexico_2015 = mexico_data[mexico_data["Year"] == 2015]
mexico_2015_life_exp = mexico_2015["Life expectancy at birth (years)"].values[0]
mexico_2000 = mexico_data[mexico_data["Year"] == 2000]
mexico_2000_life_exp = mexico_2000["Life expectancy at birth (years)"].values[0]
list_of_all_coutries_life_exp_2015.append(mexico_2015_life_exp)
list_of_all_coutries_life_exp_2000.append(mexico_2000_life_exp)

#Zimbabwe
zim_data = data[data["Country"] == "Zimbabwe"]
zim_2015 = zim_data[zim_data["Year"] == 2015]
zim_2015_life_exp = zim_2015["Life expectancy at birth (years)"].values[0]
zim_2000 = zim_data[zim_data["Year"] == 2000]
zim_2000_life_exp = zim_2000["Life expectancy at birth (years)"].values[0]
list_of_all_coutries_life_exp_2015.append(zim_2015_life_exp)
list_of_all_coutries_life_exp_2000.append(zim_2000_life_exp)

print(list_of_all_coutries_life_exp_2015)
print(list_of_all_coutries_life_exp_2000)
```

```python
plt.figure(figsize=(10, 8))
#ax = plt.subplot()
plt.bar(list_of_countries, list_of_all_coutries_life_exp_2015)
#plt.bar(list_of_countries, list_of_all_coutries_life_exp_2000)
plt.xlabel("Countries")
plt.ylabel("Life expectancy in years")
plt.title("Life expectancies in 2015")
#plt.legend()
plt.show()
plt.clf()
```

```python
#Zimbabwe's gdp development
years = zim_data["Year"].tolist()
gdp = zim_data["GDP"].tolist()

#usa's same development
years_usa = usa_data["Year"].tolist()
gdp_usa = usa_data["GDP"].tolist()

#chile's same dev
years_chile = chile_data["Year"].tolist()
gdp_chile = chile_data["GDP"].tolist()

plt.plot(years, gdp, marker='o')
#plt.plot(years_usa, gdp_usa, marker='o', color='red')
plt.plot(years, gdp_chile, marker='o', color="green")
plt.title("GDP - Chile vs Zimbabwe")
plt.show()
plt.clf()

#usa alone
plt.plot(years, gdp_usa, marker='o', color='red')
plt.title("USA's gdp")
plt.show()
plt.clf()

#zim alone
plt.plot(years, gdp, marker='o')
plt.title("Zimbabwe's gdp")
plt.show()
plt.clf()

#chile alone
plt.plot(years, gdp_chile, marker='o', color="green")
plt.title("Chile's gdp")
plt.show()
plt.clf()

#all three in a same plot
plt.plot(years, gdp_usa, marker='o', color='red', label="USA")
plt.plot(years, gdp, marker='o', label="Zimbabwe")
plt.plot(years, gdp_chile, marker='o', color="green", label="Chile")
plt.title("USA's, Zimbabwe's and Chile's' GDP comparison")
plt.show()
plt.clf()
```
